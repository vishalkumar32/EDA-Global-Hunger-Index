```python
df_ghi = pd.read_csv("eda-datas/global-hunger-index.csv")
df_underweight = pd.read_csv("eda-datas/share-of-children-underweight.csv")
df_lowweight = pd.read_csv("eda-datas/share-of-children-with-a-weight-too-low-for-their-height-wasting.csv")
df_stunting = pd.read_csv("eda-datas/share-of-children-younger-than-5-who-suffer-from-stunting.csv")
```

```python
col=[2000,2006,2012,2021]
fig = plt.figure(figsize=(15,5))

for i in range(len(col)):
    plt.subplot(1,4,i+1)
    plt.title(col[i])
    sns.distplot(df_hunger_index2,x=df_hunger_index2[col[i]])
    plt.xlim([0,75])
    plt.ylim([0,0.06])
plt.tight_layout()
plt.show()
```

```python
col=[2000,2006,2012,2021]
plt.figure(figsize=(10, 5))
for i in col:
    sns.distplot(df_hunger_index2[i],axlabel='Hunger Index')
```

```python
df_hunger_index2[df_hunger_index2[2000]>=50].sort_values(by=2000,ascending=True).plot.barh(y=2000,x='Entity',legend=False);
```
